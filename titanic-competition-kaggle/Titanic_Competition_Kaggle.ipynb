#Eksplorasi Data

### Feature Engineering Cabin

```python
#Mengubah cabin number menjadi Deck
import string
import numpy as np

def substrings_in_string(big_string, substrings):
    if not isinstance(big_string, str):
        big_string = str(big_string)

    for substring in substrings:
        if substring in big_string:
            return substring
        print(big_string)
    return np.nan
```

```python
cabin_list = ['A', 'B', 'C', 'D', 'E', 'F', 'T', 'G', 'Unknown']
train['Deck']=train['Cabin'].map(lambda x: substrings_in_string(x, cabin_list))
test['Deck']=test['Cabin'].map(lambda x: substrings_in_string(x, cabin_list))
```

### Feature Engineering SibSp and Parch into Family Size

```python
#Membuat kolom baru yaitu family_size
train['Family_Size']=train['SibSp']+train['Parch']
test['Family_Size']=test['SibSp']+test['Parch']
```

### Title from Name

```python
#Ekstrak title dari kolom 'Name'
train['Title'] = train['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test['Title'] = test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

#Lihat hasil unik
print(train['Title'].unique())
print(test['Title'].unique())
```

```python
print(train['Title'].value_counts())
print(test['Title'].value_counts())
```

```python
#Ganti title yang jarang menjadi 'Rare'
#Train
train['Title'] = train['Title'].replace(['Lady', 'Countess','Capt','Col','Don','Dr',
                                   'Major','Rev','Sir','Jonkheer'], 'Rare')
train['Title'] = train['Title'].replace(['Mlle','Ms'], 'Miss')
train['Title'] = train['Title'].replace('Mme', 'Mrs')

#Test
test['Title'] = test['Title'].replace(['Dona','Dr','Rev','Col'], 'Rare')
test['Title'] = test['Title'].replace(['Ms'], 'Miss')

print(train['Title'].value_counts())
print(test['Title'].value_counts())
```

### Ticket

```python
##Ekstrak TicketPrefix (hapus angka, ambil huruf/kode di awal)
#Train
train['TicketPrefix'] = train['Ticket'].str.replace('[\.\/]', '', regex=True) \
                                  .str.replace('\d+', '', regex=True) \
                                  .str.strip()

##Untuk tiket yang tidak punya TicketPrefix, isi 'NONE'
train['TicketPrefix'] = train['TicketPrefix'].replace('', 'NONE')

#Test
test['TicketPrefix'] = test['Ticket'].str.replace('[\.\/]', '', regex=True) \
                                  .str.replace('\d+', '', regex=True) \
                                  .str.strip()

test['TicketPrefix'] = test['TicketPrefix'].replace('', 'NONE')

print(train['TicketPrefix'].value_counts().head(10))
print(test['TicketPrefix'].value_counts().head(10))
```

### IsAlone or Not

```python
#Train
train['IsAlone'] = 0
train.loc[train['Family_Size'] == 1, 'IsAlone'] = 1

#Test
test['IsAlone'] = 0
test.loc[test['Family_Size'] == 1, 'IsAlone'] = 1
```

```python
print(train['IsAlone'].value_counts())
print(test['IsAlone'].value_counts())
```

### Labeling

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

#train
train['Sex Code'] = le.fit_transform(train['Sex'])
train['Embarked Code'] = le.fit_transform(train['Embarked'])
train['Deck Code'] = le.fit_transform(train['Deck'])
train['Title Code'] = le.fit_transform(train['Title'])
train['TicketPrefix Code'] = le.fit_transform(train['TicketPrefix'])

#test
test['Sex Code'] = le.fit_transform(test['Sex'])
test['Embarked Code'] = le.fit_transform(test['Embarked'])
test['Deck Code'] = le.fit_transform(test['Deck'])
test['Title Code'] = le.fit_transform(test['Title'])
test['TicketPrefix Code'] = le.fit_transform(test['TicketPrefix'])
```

### Preprocessing Missing Value

```python
missing_valuestrain = train.isnull().sum()
missing_valuestest = test.isnull().sum()
print('Missing Value Data Train')
print(missing_valuestrain)
print("-------------------------------")
print('-------------------------------')
print('Missing Value Data Test')
print(missing_valuestest)
```

```python
train[['Age']] = train[['Age']].fillna(train[['Age']].mean())
test[['Age','Fare']] = test[['Age','Fare']].fillna(test[['Age','Fare']].mean())
```

```python
def imputasi_acak(series):
    nilai_tersedia = series.dropna()
    return series.apply(lambda x: np.random.choice(nilai_tersedia) if pd.isnull(x) else x)

train['Embarked'] = imputasi_acak(train['Embarked'])
train['Deck'] = imputasi_acak(train['Deck'])
test['Deck'] = imputasi_acak(test['Deck'])
```

# Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, Y_train)
Y_pred = rf.predict(X_test)
```

# SVM (Support Vector Machine)

```python
from sklearn.svm import SVC

# Inisialisasi model
svm_model = SVC()

# Melatih model
svm_model.fit(X_train, Y_train)

# Melakukan prediksi
y_pred = svm_model.predict(X_test)
```

# KNN (K-Nearest Neighbor)

```python
from sklearn.neighbors import KNeighborsClassifier

# Inisialisasi model dengan k=5
knn_model = KNeighborsClassifier(n_neighbors=5)

# Melatih model
knn_model.fit(X_train, Y_train)

# Melakukan prediksi
Y_predKNN = knn_model.predict(X_test)
```

# Ada Boost

```python
from sklearn.ensemble import AdaBoostClassifier

# Inisialisasi model
ada_model = AdaBoostClassifier()

# Melatih model
ada_model.fit(X_train, Y_train)

# Melakukan prediksi
Y_predAB = ada_model.predict(X_test)
```

# XGBoost

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

xgb_model.fit(X_train, Y_train)
y_pred_xgb = xgb_model.predict(X_test)
```

# Light GBM

```python
import lightgbm as lgb

lgb_model = lgb.LGBMClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=-1,
    num_leaves=31,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

lgb_model.fit(X_train, Y_train)
y_pred_lgb = lgb_model.predict(X_test)
```

# SVM Hyperparameter

```python
from sklearn.model_selection import GridSearchCV

svm_hyper = SVC(probability=True, random_state=42)

param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 0.01, 0.001]
}

grid_svm = GridSearchCV(
    estimator=svm_hyper,
    param_grid=param_grid_svm,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_svm.fit(X_train, Y_train)
print("Best SVM Params:", grid_svm.best_params_)
print("Best SVM CV Accuracy:", grid_svm.best_score_)
```

```python
# Ambil model terbaik hasil CV
best_svm = grid_svm.best_estimator_

# Prediksi di validation set
Y_predSVM = best_svm.predict(X_test)
```

# Light GBM Hyper

```python
lgb_hyper = lgb.LGBMClassifier(random_state=42)

param_grid_lgb = {
    'n_estimators': [50, 100, 200, 500],
    'learning_rate': [0.01, 0.05, 0.1],
    'num_leaves': [31, 50, 100],
    'max_depth': [-1, 5, 10],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_lgb = GridSearchCV(
    estimator=lgb_hyper,
    param_grid=param_grid_lgb,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_lgb.fit(X_train, Y_train)

print("Best LGB Params:", grid_lgb.best_params_)
print("Best LGB CV Accuracy:", grid_lgb.best_score_)
```

# Voting Classifier XGB dan LGB

```python
from sklearn.ensemble import VotingClassifier

ensemble_model = VotingClassifier(
    estimators=[
        ('xgb', xgb.XGBClassifier(
            n_estimators=500, learning_rate=0.05, max_depth=4,
            subsample=0.8, colsample_bytree=0.8, random_state=42
        )),
        ('lgb', lgb.LGBMClassifier(
            n_estimators=500, learning_rate=0.05, max_depth=-1,
            num_leaves=31, subsample=0.8, colsample_bytree=0.8, random_state=42
        ))
    ],
    voting='soft'
)

ensemble_model.fit(X_train, Y_train)
Y_pred_ens = ensemble_model.predict(X_test)
```

```python
# Ambil model terbaik hasil CV
best_lgb = grid_lgb.best_estimator_

# Prediksi di validation set
Y_predLGB = best_lgb.predict(X_test)
```
